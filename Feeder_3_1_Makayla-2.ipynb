# Creating Your First Subset
For this project, we are looking at the number of teen births in Arkansas. Now that we have imported our dataframe as df, we can begin filtering specific columns and rows. If we want to select multiple columns, we must list them in their own nested set of square brackets.

- In this case, the data for counties in Arizona is in rows 1-47. In addition, state, county and teen births are all seperate columns. We can filter them by using the following code.

```python
df[["State","County","Teen births"]][1:47]
```

- Next, we are going to assign an easy-to-remember name to our subset. Let's call it "new_subset"

```python
new_subset=df[["State","County","Teen births"]][1:47]
```

Now, lets make sure our data appears correctly.

```python
new_subset
```

If this works, you should see the same dataset you saw when running the original code in step one.

Now that we have created our subset, we are ready to export!


# Exporting Your Subset

In order to export our subset we need to use the .to_csv() method.

The name of the file goes in parenthesis along with index=False in order to avoid the addition of an unneeded row of indices that Pandas adds. We are also going to choose a name for our new subset. Mine is going to be, "Teen_Births_in_Arkansas.csv"

With all of this together, we should have the following code.

```python
new_subset.to_csv("Teen_Births_in_Arkansas.csv",index=False)
```

The subset should show up in your content folder, which you can then download and add to the folder you created earlier for storing all the project related files.

## Congrats, you just made your first subset!!
